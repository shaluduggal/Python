### instance , class and static methods


```python
#instance , class and static methods
class Person:
    person_count=0  #static variable
    def __init__(self, name, age,address):
        self.name = name
        self.age = age
        self.address=address
        Person.person_count+=1   #static variable accessed using class name
    def display(self):
        print('Name ={} and Age {}'.format(self.name,self.age))
    
    
    @classmethod   #class method to access, manipulate static variables in the class, instance var. are not accessible here
    def display_count(cls):  #cls mandatory parameter to repesent class method used to access static var
        print('Number of Person objects = {}'.format(cls.person_count))
      
    # a static method is a general utility function neither access instance var nor static variables
    @staticmethod
    def static_method():
        return 0;
   
person1 = Person('mayank', 21,'Punjab')
person2 = Person('Priyansh', 22,'Jalandhar')
Person.display_count() # class method can be access via class name or instance
person1.display_count() 
#Person.display() #cant call static var using instance method
person1.display()
```

# overloading concept using magic methods

```python
#overloading concept using magic methods 
class Distance:
      
    # magic method to initiate object
    def __init__(self, dist):
        self.dist =dist 
          
    # print our object
    def __repr__(self):
        return 'Object: {}'.format(self.dist)
          
    def __add__(self, other):
        print('Overloaded magic method add is invoked')
        return Distance(self.dist + other.dist)
    def __mul__(self, other):
        print('Overloaded magic method add is invoked')
        return Distance(self.dist * other.dist)
    
    def __iadd__(self, other):   #overload += operator
        print('Overloaded magic method +=add is invoked')
        self.dist += other.dist
        return self
    def __ge__(self, other):
        print('Overloaded magic method greater than is invoked')
        return self.dist>=other.dist
    
    def __str__(self):  #called when oject reference is used in printing 
        return str(self.dist)

  
if __name__ == '__main__':
      
    # object creation
    distance1 = Distance(120)
    distance2 = Distance(80)
    distance3=Distance(20)
    print('Distance object has value {}'.format(distance1))
    total_distance=distance1+(distance2 *distance3) #clalling __add__ method to add two objects self -->distance1  other--->distance2
    print('Total distance covered is {}'.format(total_distance)) # called __str__
    #print('Total distance covered is {}'.format(total_distance.dist))  #execute when __str__ not defined
    distance1 += distance2 #clalling __iadd__ method to add two objects self -->distance1  other--->distance2
    print('Total distance covered is {}'.format(total_distance.dist))
    if (distance1>=distance2):
        print('Distance 1 is higher than distance 2')
    else:
        print('Distance 2 is higher than distance 2')
    
    
```

# overloading with multiple class objects


##### Magic methods in Python are the special methods that start and end with the double underscores. They are also called dunder methods. Magic methods are not meant to be invoked directly by you, but the invocation happens internally from the class on a certain action. For example, when you add two numbers using the + operator, internally, the __add__() method will be called.

```python
#overloading with multiple class objects
class WorkDays:
      
    # magic method to initiate object
    def __init__(self, days):
        self.days =days
          
    # print our object
    def __repr__(self):
        return 'Object: {}'.format(self.days)
          
    def __add__(self, other):
        print('Overloaded magic method add is invoked')
        return WorkDays(self.days + other.days)
    def __mul__(self, other):
        print('Overloaded magic method mul. is invoked in Workdays')
        return (self.days*other.per_day_salary)
    
#overloading with multiple class objects
class Emp:
      
    # magic method to initiate object
    def __init__(self, per_day_salary):
        self.per_day_salary =per_day_salary
          
    # print our object
    def __repr__(self):
        return 'Object: {}'.format(self.per_day_salary)
          
    def __mul__(self, other):
        print('Overloaded magic method mul. is invoked  in emp')
        return (self.per_day_salary*other.days)
    
e1=Emp(200)
wd1=WorkDays(20)
salary=str(e1*wd1)
print('Total salary is ',salary)  #here e1 is called object, wdi parameter  so __mul__ defined in Emp class for this operation

salary=str(wd1*e1)
print('Total salary is ',salary)  #here wdi is called object, e1 parameter  so __mul__ defined in WorkDays  class for this operation

```

# inheritance :
Capability of one class to derive or inherit the properties from another class.
benefits : Real-world relationships well.
It provides reusability of a code. We don’t have to write the same code again and again.
It is transitive in nature, which means that if class B inherits from another class A, then all the subclasses of B would automatically inherit from class A.


```python
class Person:
    def __init__(self, fname, lname, address):
        self.fname = fname
        self.lname = lname
        self.address = address
        
    def display(self):
        print("First Name: ", self.fname)
        print("Last Name: ", self.lname)
        print("Address: ", self.address)
        
class Student(Person):
    def __init__(self, fname, lname, address, age, gradYear):
        super().__init__(fname, lname, address)  #calling parent class constructor and passing parameter
        self.age = age
        self.gradYear = gradYear
        
    def display(self):
        super().display()
        print("Age: ", self.age)
        print("Graduation Year: ", self.gradYear)

# person object
per = Person("Priya", "Tondon", "Jalandhar") 
per.display()
print("===========================================")
std = Student("Rekha", "Verma", "WC 65,Lawrence Road, Amritsar",20,2001)
std.display()
```

##### --------Composition
In composition one class acts as a container of the other class (contents). If you destroy the container there is no existence of contents. That means if the container class creates an object or hold an object of contents.
Composition established “has-a” relationship between objects. In below code, you can see that the class person is creating a heart object. So, person is the owner of the heart object.

# Demonstrating composition


```python
#Demonstrating composition

class Heart:
    def __init__(self, heartValves):
        self.heartValves = heartValves
        
    def display(self):
        return self.heartValves
    
class Person:
    def __init__(self, fname, lname, address, heartValves):
        self.fname = fname
        self.lname = lname
        self.address = address
        self.heartValves = heartValves
        self.heartObject = Heart(self.heartValves)   # Composition
        
    def display(self):
        print("First Name: ", self.fname)
        print("Last Name: ", self.lname)
        print("Address: ", self.address)
        print("No of Heart Valves: ", self.heartObject.display())


p = Person("Sia", "Mehta", "Model town , Jalandhar", 4)
p.display()
```

###### # Inheritance supported in Python
Single Inheritance: Single inheritance enables a derived class to inherit properties from a single parent class, 
    thus enabling code reusability and the addition of new features to existing code.
 

Multiple Inheritance: When a class can be derived from more than one base class this type of inheritance is called 
        multiple inheritance. In multiple inheritance, all the features of the base classes are inherited into the
        derived class. 

        
Multilevel Inheritance 
In multilevel inheritance, features of the base class and the derived class are further inherited into the new
derived class. This is similar to a relationship representing a child and grandfather. 



Hierarchical Inheritance: When more than one derived classes are created from a single base this type of 
    inheritance is called hierarchical inheritance. In this program, we have a parent (base) class and 
    two child (derived) classes.
 

Hybrid Inheritance: Inheritance consisting of multiple types of inheritance is called hybrid inheritance.



# single inheritance

```python
class Parent:
    def __init__(self,name):
        self.parentname=name    
    def display(self):
        print("Parent name is {}".format(self.parentname))
class Child(Parent):
    def __init__(self,childname,parentname):
        super().__init__(parentname)
        self.childname=childname   
    def display(self):
        super().display()
        print("child name is {}".format(self.childname))

o2=Parent('Meenu')
o2.display()
print('*'*20)
o1= Child('Reeta','Ajay')
o1.display()
```

# Mutiple inheritance, Resolving ambiguity problem of same function name in parent classes called in child class 

```python
class Father:
    def __init__(self,name):
        self.father=name    
    def display(self):
        print("Father name is {}".format(self.father))
class Mother:
    def __init__(self,name):
        self.mother=name    
    def display(self):
        print("Mother name is {}".format(self.mother))


        
class Child1(Father,Mother):  #multiple inheritance -- resolve "display() call" by checking the ordering of parents
    def __init__(self,childname,father,mother):
        
        Father.__init__(self,father)
        Mother.__init__(self,mother)
        self.childname=childname   
    
print('Child 1')
o1=Child('Reeta','Ajay','Manju')
o1.display()


#after changing order of parents
class Child2(Mother,Father):  #multiple inheritance -- resolve "display() call" by checking the ordering of parents
    def __init__(self,childname,mother,father):
        Mother.__init__(self,mother)
        self.childname=childname   

print('_'*20)
print(''*20)
print(''*20)
print('Child 2')
o2=Child2('Reeta','Manju','Ajay')
o2.display()




#after overwriting display in child both parent display is called
class Child3(Mother,Father):  #multiple inheritance -- resolve "display() call" by checking the ordering of parents
    def __init__(self,childname,mother,father):
        Mother.__init__(self,mother)
        Father.__init__(self,father)
        self.childname=childname   
    def display(self):
        Father.display(self)
        Mother.display(self)
        print("child name is {}".format(self.childname))


print('_'*20)
print(''*20)
print(''*20)
print('Child 3')
o3=Child3('Reeta','Manju','Ajay')
o3.display()
```

# Method Resolution Order MRO
MRO is a concept used in inheritance. It is the order in which a method is searched for in a classes hierarchy and is especially useful in Python because Python supports multiple inheritance.

In Python, the MRO is from bottom to top and left to right. This means that, first, the method is searched in the class of the object. If it’s not found, it is searched in the immediate super class. In the case of multiple super classes, it is searched left to right, in the order by which was declared by the developer
Python follows a depth-first lookup order
Python(3.x & 2.2) we are bound to use only new classes. New style classes are the ones whose first parent inherits from Python root ‘object’ class.


## Old style classes use DLR ( depth-first left to right algorithm). New style classes use C3 Linearization algorithm for method resolution
